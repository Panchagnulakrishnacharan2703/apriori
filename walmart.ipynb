```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
walmart = pd.read_excel('/content/Online Retail(1).xlsx')
walmart.head()
```

```python
desc = walmart

diction = {
    'light' = ['light','lantern'],
    ''
}
```

```python
walmart.columns # to know the atributes 
 
```

```python
walmart.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
```

```python
#data cleaning ie preprocessing

walmart['Description'] = walmart['Description'].str.strip()
walmart.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
walmart['InvoiceNo'] = walmart['InvoiceNo'].astype('str')
walmart = walmart[~walmart['InvoiceNo'].str.contains('C')]
walmart.isnull().sum()
```

```python

France = (walmart[walmart['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
 

UK = (walmart[walmart['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
  

Por = (walmart[walmart['Country'] =="Portugal"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
  
Sweden = (walmart[walmart['Country'] =="Sweden"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
France
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
  
# Encoding the datasets
basket_encoded = France.applymap(hot_encode)
France = basket_encoded
  
basket_encoded =UK.applymap(hot_encode)
UK = basket_encoded
  
basket_encoded = Por.applymap(hot_encode)
Por = basket_encoded
  
basket_encoded = Sweden.applymap(hot_encode)
Sweden = basket_encoded
```

```python
frq_items = apriori(France, min_support = 0.05, use_colnames = True)
  

rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head())
```

```python
def apriori_assoc(country:str,min_support:int):
  frq_items = apriori(country, min_support = min_support, use_colnames = True)
  rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
  rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
  return rules.head()
```
